```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
import pandas as pd
from bokeh.models import ColumnDataSource
output_notebook()
data = pd.read_csv('C:/Users/khamz/Downloads/voting_data_geo_eng.csv')
(data)[:10]
```

```python
candidates = ['Baburin', 'Grudinin', 'Zhirinovskiy ', 'Putin', 'Sobchak', 'Suraikin', 'Titov', 'Yavlinskiy']

p = figure(x_range=candidates, plot_height=500, title="2018 Presidential Voting Results in Russia in %")

S = sum(data['Number of valid ballot papers'])

p.vbar(x=candidates, top=[100*sum(data['Baburin Sergei Nikolaevich'])/S, 100*sum(data['Grudinin Pavel Nikolaevich'])/S,
                          100*sum(data['Zhirinovskiy Vladimir Volfovich'])/S, 100*sum(data['Putin Vladimir Vladimirovich'])/S,
                          100*sum(data['Sobchak Ksenia Anatolyevna'])/S, 100*sum(data['Suraikin Maksim Aleksandrovich'])/S,
                          100*sum(data['Titov Boris Yurievich'])/S, 100*sum(data['Yavlinskiy Gregory Alekseivich'])/S], width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
from bokeh.palettes import GnBu3, OrRd3

moscow = data[data.region_name == 'gorod Moskva']
bashkortostan = data[data.region_name == 'Respublika Baskortostan']
volgograd = data[data.region_name == 'Vologodskaa oblast_']

M = sum(moscow['Number of valid ballot papers'])
B = sum(bashkortostan['Number of valid ballot papers'])
V = sum(volgograd['Number of valid ballot papers'])

regions = ['Moscow City', 'Bashkortostan Republic', 'Volgograd Region']

votes = {'candidates' : candidates,
           'Moscow City' :[100*sum(moscow['Baburin Sergei Nikolaevich'])/M, 100*sum(moscow['Grudinin Pavel Nikolaevich'])/M,
                           100*sum(moscow['Zhirinovskiy Vladimir Volfovich'])/M, 100*sum(moscow['Putin Vladimir Vladimirovich'])/M,
                           100*sum(moscow['Sobchak Ksenia Anatolyevna'])/M, 100*sum(moscow['Suraikin Maksim Aleksandrovich'])/M,
                           100*sum(moscow['Titov Boris Yurievich'])/M, 100*sum(moscow['Yavlinskiy Gregory Alekseivich'])/M],
                                   
           'Bashkortostan Republic' :[100*sum(bashkortostan['Baburin Sergei Nikolaevich'])/B, 100*sum(bashkortostan['Grudinin Pavel Nikolaevich'])/B,
                                      100*sum(bashkortostan['Zhirinovskiy Vladimir Volfovich'])/B, 100*sum(bashkortostan['Putin Vladimir Vladimirovich'])/B,
                                      100*sum(bashkortostan['Sobchak Ksenia Anatolyevna'])/B, 100*sum(bashkortostan['Suraikin Maksim Aleksandrovich'])/B,
                                      100*sum(bashkortostan['Titov Boris Yurievich'])/B, 100*sum(bashkortostan['Yavlinskiy Gregory Alekseivich'])/B],
                                              
           'Volgograd Region' :[100*sum(volgograd['Baburin Sergei Nikolaevich'])/V, 100*sum(volgograd['Grudinin Pavel Nikolaevich'])/V,
                                100*sum(volgograd['Zhirinovskiy Vladimir Volfovich'])/V, 100*sum(volgograd['Putin Vladimir Vladimirovich'])/V,
                                100*sum(volgograd['Sobchak Ksenia Anatolyevna'])/V, 100*sum(volgograd['Suraikin Maksim Aleksandrovich'])/V,
                                100*sum(volgograd['Titov Boris Yurievich'])/V, 100*sum(volgograd['Yavlinskiy Gregory Alekseivich'])/V]}

p = figure(y_range=candidates, plot_height=500, title="2018 Presidential Voting Results in Moscow, Bashkortostan, Vologograd Region")

p.hbar_stack(regions, y='candidates', height=0.9, color=OrRd3, source=ColumnDataSource(votes),
             legend=["%s votes" % x for x in regions])

p.y_range.range_padding = 0.1
p.ygrid.grid_line_color = None
p.legend.location = "top_right"

show(p)
```

```python
from bokeh.models import FactorRange
from bokeh.transform import factor_cmap

candidates = ['Baburin', 'Grudinin', 'Zhirinovskiy ', 'Putin', 'Sobchak', 'Suraikin', 'Titov', 'Yavlinskiy']
regions = ['Moscow', 'Bashkortostan', 'Volgograd']

data = {'candidates' : candidates,
           'Moscow' :[100*sum(moscow['Baburin Sergei Nikolaevich'])/M, 100*sum(moscow['Grudinin Pavel Nikolaevich'])/M,
                           100*sum(moscow['Zhirinovskiy Vladimir Volfovich'])/M, 100*sum(moscow['Putin Vladimir Vladimirovich'])/M,
                           100*sum(moscow['Sobchak Ksenia Anatolyevna'])/M, 100*sum(moscow['Suraikin Maksim Aleksandrovich'])/M,
                           100*sum(moscow['Titov Boris Yurievich'])/M, 100*sum(moscow['Yavlinskiy Gregory Alekseivich'])/M],
                                   
           'Bashkortostan' :[100*sum(bashkortostan['Baburin Sergei Nikolaevich'])/B, 100*sum(bashkortostan['Grudinin Pavel Nikolaevich'])/B,
                                      100*sum(bashkortostan['Zhirinovskiy Vladimir Volfovich'])/B, 100*sum(bashkortostan['Putin Vladimir Vladimirovich'])/B,
                                      100*sum(bashkortostan['Sobchak Ksenia Anatolyevna'])/B, 100*sum(bashkortostan['Suraikin Maksim Aleksandrovich'])/B,
                                      100*sum(bashkortostan['Titov Boris Yurievich'])/B, 100*sum(bashkortostan['Yavlinskiy Gregory Alekseivich'])/B],
                                              
           'Volgograd' :[100*sum(volgograd['Baburin Sergei Nikolaevich'])/V, 100*sum(volgograd['Grudinin Pavel Nikolaevich'])/V,
                                100*sum(volgograd['Zhirinovskiy Vladimir Volfovich'])/V, 100*sum(volgograd['Putin Vladimir Vladimirovich'])/V,
                                100*sum(volgograd['Sobchak Ksenia Anatolyevna'])/V, 100*sum(volgograd['Suraikin Maksim Aleksandrovich'])/V,
                                100*sum(volgograd['Titov Boris Yurievich'])/V, 100*sum(volgograd['Yavlinskiy Gregory Alekseivich'])/V]}

x = [ (candidate, region) for candidate in candidates for region in regions ]
counts = sum(zip(data['Moscow'], data['Bashkortostan'], data['Volgograd']), ()) 

source = ColumnDataSource(data=dict(x=x, counts=counts))

p = figure(x_range=FactorRange(*x), plot_height=500, title="2018 Presidential Voting Results in Russia by Moscow, Bashkortostan, and Vologograd")

p.vbar(x='x', top='counts', width=0.9, source=source, line_color="black", fill_color=factor_cmap('x', palette=['white', 'navy', 'firebrick'], factors=regions, start=1, end=2))
p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None

show(p)
```
